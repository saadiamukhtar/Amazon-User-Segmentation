```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv('Amazon_model.csv')
X= dataset.iloc[:,[3,4]].values
print(X)
```

```python
from sklearn.cluster import KMeans
wcss= []
for i in range(1,11):
  kmeans= KMeans(n_clusters=i, init="k-means++", random_state=21)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("Wcss via Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS value")

plt.show()
```

```python
  kmeans= KMeans(n_clusters=4, init="k-means++", random_state=21)
  ymeans= kmeans.fit_predict(X)
```

```python
print(ymeans)
```

```python
plt.scatter(X[ymeans==0 ,0], X[ymeans==0,1], s=100, c="magenta", label="cluster1")
plt.scatter(X[ymeans==1 ,0],X[ymeans==1,1], s=100, c="cyan", label="cluster2")
plt.scatter(X[ymeans==2 ,0],X[ymeans==2,1], s=100,  c="blue", label="cluster3")
plt.scatter(X[ymeans==3 ,0],X[ymeans==3,1], s=100,  c="red", label="cluster4")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=300, c="black", label="centroids")
plt.title("Cluster of amazon users")
plt.xlabel("Annual Income in INR")
plt.ylabel("Purchase rating")
plt.legend()
plt.show()
```
